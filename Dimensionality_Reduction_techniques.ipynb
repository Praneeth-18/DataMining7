# **PCA**

PCA is a linear dimensionality reduction technique that transforms the data into a new coordinate system, where the greatest variance comes to lie on the first few principal components.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
```

```python
# Load the dataset
digits = load_digits()
X_digits, y_digits = digits.data, digits.target
```

```python
# Applying PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(X_digits)
```

```python
# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_digits, alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.colorbar(label='Digit Label')
plt.title('PCA of Digits Dataset')
plt.show()
```

# **Singular Value Decomposition (SVD)**
SVD is another popular linear dimensionality reduction method, often used in the context of data processing and feature extraction. SVD is a matrix factorization technique that decomposes a matrix into three other matrices. It is used in many areas including dimensionality reduction, signal processing, and statistics.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD
from sklearn.datasets import load_digits
```

```python
# Load the dataset (if not already loaded)
digits = load_digits()
X_digits, y_digits = digits.data, digits.target
```

```python
# Applying Truncated SVD
svd = TruncatedSVD(n_components=2)
X_svd = svd.fit_transform(X_digits)
```

```python
# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y_digits, alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.xlabel('First singular vector')
plt.ylabel('Second singular vector')
plt.colorbar(label='Digit Label')
plt.title('SVD of Digits Dataset')
plt.show()
```

# **Multidimensional Scaling (MDS)**
MDS is a non-linear dimensionality reduction technique that aims to place each object in N-dimensional space such that the distances between objects are preserved as well as possible.

It's particularly useful in visualizing similarity or dissimilarity data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from sklearn.datasets import load_digits
```

```python
# Load the dataset (if not already loaded)
digits = load_digits()
X_digits, y_digits = digits.data, digits.target
```

```python
# Applying MDS
mds = MDS(n_components=2, random_state=42)
X_mds = mds.fit_transform(X_digits)
```

```python
# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_mds[:, 0], X_mds[:, 1], c=y_digits, alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.xlabel('First MDS dimension')
plt.ylabel('Second MDS dimension')
plt.colorbar(label='Digit Label')
plt.title('MDS of Digits Dataset')
plt.show()
```

# **ISOMap**
ISOMap is a technique used for manifold learning. It is based on computing the geodesic distances among the points, which is effective for unfolding twisted and curved manifolds.

It's particularly useful for datasets where linear methods like PCA and SVD fail to capture the intrinsic geometry.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import Isomap
from sklearn.datasets import load_digits
```

```python
# Load the dataset (if not already loaded)
digits = load_digits()
X_digits, y_digits = digits.data, digits.target
```

```python
# Applying ISOMap
isomap = Isomap(n_components=2)
X_isomap = isomap.fit_transform(X_digits)
```

```python
# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_isomap[:, 0], X_isomap[:, 1], c=y_digits, alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.xlabel('First ISOMap dimension')
plt.ylabel('Second ISOMap dimension')
plt.colorbar(label='Digit Label')
plt.title('ISOMap of Digits Dataset')
plt.show()
```

# **Locally Linear Embedding (LLE)**
LLE works by first measuring how each data point linearly relates to its nearest neighbors, and then seeking a lower-dimensional representation of the data that preserves these local relationships as much as possible.

It's effective for datasets where global linear methods like PCA fail to capture complex, local structures.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.datasets import load_digits
```

```python
# Load the dataset (if not already loaded)
digits = load_digits()
X_digits, y_digits = digits.data, digits.target
```

```python
# Applying LLE
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10, random_state=42)
X_lle = lle.fit_transform(X_digits)
```

```python
# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=y_digits, alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.xlabel('First LLE dimension')
plt.ylabel('Second LLE dimension')
plt.colorbar(label='Digit Label')
plt.title('LLE of Digits Dataset')
plt.show()
```

# **t-SNE**
t-SNE is effective in capturing the local structure of high-dimensional data and revealing clusters in the data.

It's especially popular for visualizing complex datasets with inherent clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
```

```python
# Load the dataset (if not already loaded)
digits = load_digits()
X_digits, y_digits = digits.data, digits.target
```

```python
# Applying t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_digits)
```

```python
# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_digits, alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.colorbar(label='Digit Label')
plt.title('t-SNE of Digits Dataset')
plt.show()
```

# **UMAP (Uniform Manifold Approximation and Projection)**
UMAP is a manifold learning technique ideal for embedding high-dimensional data into lower-dimensional spaces.

It is similar to t-SNE in its ability to reveal the local and global structure of data but often works faster and scales better to larger datasets.

```python
!pip install umap-learn
import umap
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
# Load the dataset (if not already loaded)
digits = load_digits()
X_digits, y_digits = digits.data, digits.target
```

```python
# Applying UMAP
umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X_digits)
```

```python
# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y_digits, alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.xlabel('UMAP feature 1')
plt.ylabel('UMAP feature 2')
plt.colorbar(label='Digit Label')
plt.title('UMAP of Digits Dataset')
plt.show()
```
